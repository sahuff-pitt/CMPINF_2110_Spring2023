## Spotify Songs

The Spotify Songs data set also comes from TidyTuesday:

https://github.com/rfordatascience/tidytuesday/tree/master/data/2020/2020-01-21

```python
songs_url = 'https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2020/2020-01-21/spotify_songs.csv'
```

```python
songs = pd.read_csv( songs_url )
```

```python
songs.info()
```

```python
songs.nunique()
```

```python
songs.isna().sum()
```

```python
songs.track_id.value_counts()
```

```python
songs.track_id.value_counts().value_counts()
```

```python
songs.loc[ songs.track_id == '7BKLCZ1jbUBVqRi2FVlTVw', : ]
```

One row in the `songs` data set corresponds to is *maybe* the **combination** of a `track_id` and a `playlist_id`. 

Other ways to perform the filter operation.

```python
songs.loc[ songs.track_id.isin(['7BKLCZ1jbUBVqRi2FVlTVw']), : ]
```

```python
songs.track_id.value_counts()[ songs.track_id.value_counts() == 10 ].index.to_list()
```

```python
songs.loc[ songs.track_id.isin(songs.track_id.value_counts()[ songs.track_id.value_counts() == 10 ].index.to_list()) , :]
```

Could identify the songs with 9 rows in the original data set...

```python
songs.loc[ songs.track_id.isin(songs.track_id.value_counts()[ songs.track_id.value_counts() == 9 ].index.to_list()) , :]
```

Check to see the number of rows per **combination** of `track_id` and `playlist_id`.

```python
songs.groupby(['track_id', 'playlist_id']).size().reset_index(name='num_rows')
```

```python
songs.groupby(['track_id', 'playlist_id']).size().reset_index(name='num_rows').num_rows.value_counts()
```
